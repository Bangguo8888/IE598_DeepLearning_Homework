```python
import re
import matplotlib.pyplot as plt
```

```python
# define the function that can read the Rscore and KL divergence from the output file
def get_data(filename):
    with open(filename,'r') as f:
        texts = f.read()

    lines = texts.split('\n')
    episodes = []
    scores = []
    kl = []
    for line in lines:
        score = re.search('RScore:\s+(-?\d+.\d+) RPPS',line)
        if score:
            scores.append(float(score.groups(1)[0]))
            episode = re.search('Episode:\s+(\d+) Score:',line)
            episodes.append(int(episode.groups(1)[0]))
        else:
            kl_line = re.search('KL value is (\d+.\d+)',line)
            if kl_line:
                kl.append(float(kl_line.groups(1)[0]))
    return episodes,scores,kl
```

```python
# get the episodes, scores, kl values
filename = 'result1.txt'
episodes,scores,kl = get_data(filename)
```

## The Rscores

```python
plt.figure(figsize = (20,8))
plt.plot(episodes,scores)
plt.xlabel('Episodes')
plt.ylabel('Rscores')
plt.title('Episodes V.S. Rscores')
plt.show()
```
