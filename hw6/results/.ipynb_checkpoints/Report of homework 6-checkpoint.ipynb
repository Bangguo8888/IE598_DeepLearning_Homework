## Part 1

```python
import re
import matplotlib.pyplot as plt
def get_data(filename):
    with open(filename,'r') as f:
        texts = f.read()
        
    lines = texts.split('\n')
    episodes = []
    scores = []
    for line in lines:
        score = re.search('RScore:\s+(-?\d+.\d+) RPPS',line)
        if score:
            scores.append(float(score.groups(1)[0]))
            episode = re.search('Episode:\s+(\d+) Score:',line)
            episodes.append(int(episode.groups(1)[0]))
    return episodes,scores
```

```python
filename1 = './history1/beta0.00.bw.out'
filename2 = './history1/beta0.01.bw.out'
filename3 = './history1/beta1.00.bw.out'
episodes1,scores1 = get_data(filename1)
episodes2,scores2 = get_data(filename2)
episodes3,scores3 = get_data(filename3)
```

```python
plt.figure(figsize = (12,8))
plt.plot(episodes1,scores1,label="beta = 0.00")
plt.plot(episodes2,scores2,label="beta = 0.01")
plt.plot(episodes3,scores3,label="beta = 1.00")
plt.xlabel('Episodes')
plt.ylabel('Rscores')
plt.legend(loc='upper left')
plt.title('Episodes V.S. Rscores')
plt.show()
```

From the results, we can see that,when $\beta = 0.01$, the performance is best. The Rscores are probably 1750 when it converges. The performance when $\beta = 0$ is worse than that when $\beta = 0.01$ but is better than that $\beta = 1.00$ 
 
 
And we can also see that, when $\beta = 1$, the Rscores converges very quickly, probably after 1000 episodes and the converged Rscores is pretty low, which is about 130. While when $\beta = 0.00$, it needs about 10000 episodes to become stable and the Rscores are about 900.While when $\beta = 0.01$, the Rscores increases exponentially before 15000 episodes.After that it becomes stable and the Rscores are about 1750.

## Part 2

```python
filename4 = './history2/lambda0.bw.out'
filename5 = './history2/lambda0.9.bw.out'
filename6 = './history2/lambda1.bw.out'
episodes4,scores4 = get_data(filename4)
episodes5,scores5 = get_data(filename5)
episodes6,scores6 = get_data(filename6)
```

```python
plt.figure(figsize = (12,8))
plt.plot(episodes4,scores4,label="lambda = 0.00")
plt.plot(episodes5,scores5,label="lambda = 0.90")
plt.plot(episodes6,scores6,label="lambda = 1.00")
plt.xlabel('Episodes')
plt.ylabel('Rscores')
plt.legend(loc='upper left')
plt.title('Episodes V.S. Rscores')
plt.show()
```

From the results, we can see that,when $\lambda = 1$, the performance is best. When $\lambda = 0.9$, the performance is very similar to or slighly worse than the performance when $\lambda = 1$. When $\lambda = 0$,the performance is worst. 
 
 
And we can also see that, before 15000 episodes,the training Rscores gradually increase, though they are not stable,with some episodes have low Rscores. After 15000 episodes, the Rscores become stable and converge.
